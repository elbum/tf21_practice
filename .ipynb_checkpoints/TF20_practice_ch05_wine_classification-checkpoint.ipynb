```python
try:
  %tensorflow_version 2.x
except Exception:
  pass
import tensorflow as tf
```

```python
import pandas as pd
red = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv',sep=';')
```

```python
red
```

```python
white = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv',sep=';')
white
```

```python
red['type']=0
red.head()
```

```python
white['type']=1
white.head()
```

```python
wine = pd.concat([red,white])
```

```python
wine.head()
```

```python
wine.describe()
```

```python
import matplotlib.pyplot as plt
plt.hist(wine['type'])
plt.xticks([0,1])
plt.show()
print(wine['type'].value_counts())
```

```python
print(wine.info())
```

```python
wine_norm = (wine-wine.min()) / (wine.max()-wine.min())
wine_norm.head()
```

```python
wine.head()
```

```python
wine.describe()
```

```python
import numpy as np
wine_shuffle = wine_norm.sample(frac=1)
wine_shuffle.head()
```

```python
wine_np = wine_shuffle.to_numpy()
wine_np[:10]
```

```python
train_idx = int(len(wine_np)*0.8)
train_idx
```

```python
len(wine_np)
```

```python
train_X, train_Y = wine_np[:train_idx,:-1],wine_np[:train_idx,-1]
test_X , test_Y = wine_np[train_idx:,:-1],wine_np[train_idx:,-1]
train_X
train_Y
```

```python
train_Y.shape
```

```python
# doesn't work for 'sparse_categorical_crossentropy'

#train_Y = tf.keras.utils.to_categorical(train_Y,num_classes=2)
#test_Y = tf.keras.utils.to_categorical(test_Y,num_classes=2)
#train_Y[0]
```

```python
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=48,activation='relu',input_shape=(12,)),
                             tf.keras.layers.Dense(units=24,activation='relu'),
                             tf.keras.layers.Dense(units=12,activation='relu'),
                             tf.keras.layers.Dense(units=2,activation='softmax')
                                                   
])
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.07),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
type(train_Y)
```

```python
type(train_X.shape)
```

```python
train_X=train_X.to_numpy_array()
```

```python
history=model.fit(train_X,train_Y,epochs=25,batch_size=32,validation_split=0.25)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,5))
plt.subplot(1,2,1)
plt.plot(history.history['loss'],'b-',label='loss')
plt.plot(history.history['val_loss'],'r--',label='val_loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'],'g--',label='accuracy')
plt.plot(history.history['val_accuracy'],'k--',label='val_accuracy')
plt.xlabel('Epoch')
plt.ylim(0.7,1)
plt.legend()

plt.show()
```

```python
del wine['quality']
wine_norm=(wine-wine.min()) / (wine.max() - wine.min())
wine_shuffle = wine_norm.sample(frac=1)
wine_np = wine_shuffle.to_numpy()
wine_np
```

```python
train_idx = int(len(wine_np)*0.8)
```

```python
train_X , train_Y = wine_np[:train_idx,:-1],wine_np[:train_idx,-1]
train_X
```

```python
train_Y
```

```python
train_Y[:100]
```

```python
test_X , test_Y = wine_np[train_idx:,:-1],wine_np[train_idx:,-1]
```

```python
test_X
```

```python
test_Y
```

```python
y_binary = tf.keras.utils.to_categorical(train_Y,num_classes=3)
y_binary[:20]
```

```python
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=48,activation='relu',input_shape=(12,)),
                             tf.keras.layers.Dense(units=24,activation='relu'),
                             tf.keras.layers.Dense(units=12,activation='relu'),
                             tf.keras.layers.Dense(units=3,activation='softmax')
])
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.07),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(train_X,y_binary,batch_size=32,epochs=2500,validation_split=0.25)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,15))
plt.subplot(2,1,1)
plt.plot(history.history['loss'],'b-',label='loss')
plt.plot(history.history['val_loss'],'r--',label='val_loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(2,1,2)
plt.plot(history.history['accuracy'],'g--',label='accuracy')
plt.plot(history.history['val_accuracy'],'k--',label='val_accuracy')
plt.xlabel('Epoch')
plt.ylim(0.7,1)
plt.legend()

plt.show()
```
